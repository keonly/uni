```python
def toTexture(image, gt_image, reference, mesh):
    texture_img = np.zeros((image_height, image_width, 3), np.uint8)
    texture_img_orig = cv2.imread(gt_image)
    texture_result = texture_img_orig.copy()

    for i, triangle in enumerate(triangles):

        vertex_idx1, vertex_idx2, vertex_idx3 = triangle
        r_vertex1, r_vertex2, r_vertex3 = reference[vertex_idx1], reference[vertex_idx2], reference[vertex_idx3]
        c_vertex1, c_vertex2, c_vertex3 = mesh[vertex_idx1], mesh[vertex_idx2], mesh[vertex_idx3]

        r_x1, r_y1 = r_vertex1
        r_x2, r_y2 = r_vertex2
        r_x3, r_y3 = r_vertex3

        src_tri = np.float32([[[r_x1, r_y1], [r_x2, r_y2], [r_x3, r_y3]]])
        t_tri = np.float32([[c_vertex1, c_vertex2, c_vertex3]])

        r1 = cv2.boundingRect(src_tri)
        r2 = cv2.boundingRect(t_tri)

        tri1cropped = []
        tri2cropped = []

        for i in range(0, 3):
            tri1cropped.append(((src_tri[0, i, 0] - r1[0]), (src_tri[0, i, 1] - r1[1])))
            tri2cropped.append(((t_tri[0, i, 0] - r2[0]), (t_tri[0, i, 1] - r2[1])))

        img1Cropped = image[r1[1]:r1[1] + r1[3], r1[0]:r1[0] + r1[2]]

        warpMat = cv2.getAffineTransform(np.float32(tri1cropped), np.float32(tri2cropped))

        img2Cropped = cv2.warpAffine(img1Cropped, warpMat, (r2[2], r2[3]), None, flags=cv2.INTER_AREA, borderMode=cv2.BORDER_REFLECT)

        mask = np.zeros((r2[3], r2[2], 3), dtype=np.float32)

        cv2.fillConvexPoly(mask, np.int32(tri2cropped), (1.0, 1.0, 1.0), 16, 0)

        img2Cropped = img2Cropped * mask

        for i in range(r2[1], r2[1]+r2[3]):
            for j in range(r2[0], r2[0]+r2[2]):
                if img2Cropped[i - r2[1], j-r2[0],0] != 0 and img2Cropped[i - r2[1], j-r2[0],1] != 0 and img2Cropped[i - r2[1], j-r2[0],0] != 0:
                    if texture_img[i, j, 0] == 0 and texture_img[i, j, 1] == 0 and texture_img[i, j, 2] == 0:
                        texture_img[i, j] = img2Cropped[i-r2[1], j-r2[0]]
                    else:
                        texture_img[i, j] = (texture_img[i, j] + img2Cropped[i-r2[1],j-r2[0]]) * 0.5
    
#     texture_img = sr.upsample(texture_img)
#     texture_img = cv2.resize(texture_img, (image_height, image_width))
    
    texture_img_gray = cv2.cvtColor(texture_img, cv2.COLOR_BGR2GRAY)
    th, texture_th = cv2.threshold(texture_img_gray, 30, 50, cv2.THRESH_BINARY_INV)
    
    texture_th_floodfill = texture_th.copy()
    
    mask = np.zeros((image_height + 2, image_width + 2), np.uint8)
    
    cv2.floodFill(texture_th_floodfill, mask, (0,0), 255)
    
    texture_th_floodfill_inv = cv2.bitwise_not(texture_th_floodfill)
    
    im_out = texture_th | texture_th_floodfill_inv
    
    index_texture = np.array(list(zip(*np.where(im_out == 255))))
    
    texture_result[index_texture[:,0], index_texture[:,1]] = texture_img[index_texture[:,0], index_texture[:,1]]
    
    texture_result = cv2.seamlessClone(texture_result, texture_img_orig, im_out, (image_height//2, image_width // 2), cv2.NORMAL_CLONE)
    
#     index_texture = np.array(list(zip(*np.where(im_out == 255))))
    
#     orig_img_face = np.zeros((image_height, image_width, 3), np.uint8)
#     texture_img_face = np.zeros((image_height, image_width, 3), np.uint8)
    
#     orig_img_face[index_texture[:,0], index_texture[:,1]] = texture_img_orig[index_texture[:,0], index_texture[:,1]]
#     texture_img_face[index_texture[:,0], index_texture[:,1]] = texture_result_tmp[index_texture[:,0], index_texture[:,1]]
    
#     matched = exposure.match_histograms(texture_img_face, orig_img_face, multichannel=True)
#     texture_result[index_texture[:,0], index_texture[:,1]] = matched[index_texture[:,0], index_texture[:,1]]
    
    
#     for i in range(texture_img_orig.shape[0]):
#         for j in range(texture_img_orig.shape[1]):
#             if im_out[i, j] == 255:
#                 texture_result[i,j] = texture_img[i,j]
                
#     texture_orig_gray = cv2.cvtColor(texture_img_orig, cv2.COLOR_BGR2GRAY)

#     lapacian_result = cv2.Laplacian(texture_orig_gray, cv2.CV_64F)
#     lapacian_result = np.uint8(np.absolute(lapacian_result))
    
#     lapacian_texture = cv2.Laplacian(texture_img_gray, cv2.CV_64F)
#     lapacian_texture = np.uint8(np.absolute(lapacian_texture))
    
# #     lapacian_texture -= lapacian_result
    
#     _, lapacian_texture_th = cv2.threshold(lapacian_texture, 100, 255, cv2.THRESH_BINARY)
#     _, lapacian_result_th = cv2.threshold(lapacian_result, 5, 255, cv2.THRESH_BINARY)
    
#     index_texture = np.array(list(zip(*np.where(lapacian_texture_th == 255))))
#     index_result = np.array(list(zip(*np.where(lapacian_result_th == 255))))
    
#     texture_set = set([tuple(x) for x in index_texture])
#     result_set = set([tuple(x) for x in index_result])
#     index_final = np.array([x for x in texture_set & result_set])

#     texture_result[index_final[:,0], index_final[:,1]] = texture_img_orig[index_final[:,0], index_final[:,1]]

    return texture_result
```
